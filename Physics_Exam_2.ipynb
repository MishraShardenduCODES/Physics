```python
def fact(a):
  if(a == 0 or a == 1):
    return 1
  else:
    return a*fact(a-1)
```

```python
def fibo(a):
  if(a == 0 or a == 1):
    return a
  else:
    return fibo(a-1) + fibo(a-2)
```

```python
import math as mat
import numpy as np

def prime(a):
  if(a == 0 or a == 1):
    print("Neither Prime nor Composite")
  else:
    b = np.sqrt(a)
    c = mat.floor(b)
    stpr = 0
    for i in range(2,c+1):
      if(a % i == 0 ):
        stpr = 1
        break
    if(stpr):
      print("It's a composite number");
    else:
      print("It's a prime number")
```
