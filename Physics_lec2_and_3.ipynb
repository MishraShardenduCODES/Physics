```python
import numpy as np
import pylab as py

# Define x array
x = np.linspace(0, 100, 1000)

# Define y and y1 based on x
y = np.linspace(-0.4, +0.4, 1000)
y1 = (x**4) * np.exp(-2*x)
y2 = (x**2) * np.exp(-x) * np.sin(x**2)**2
```

```python
py.xlabel("Values")
py.ylabel("Function")
py.xlim(0,20)
py.title("Maths Function")
py.plot(x,y1,"y-",x,y2,"g-")
```
