```python
import matplotlib.pyplot as plt
import numpy as np

Radius = 40.0
radius = 10.0
wplanet = 2.0
wmoon = 14.0
time = np.arange(0., 6.4, 0.02)  # Adjusted time range for multiple loops

x0 = Radius * np.cos(wplanet * time)
y0 = Radius * np.sin(wplanet * time)

x = Radius * np.cos(wplanet * time) + radius * np.cos(wmoon * time)
y = Radius * np.sin(wplanet * time) + radius * np.sin(wmoon * time)

plt.plot(x, y, label='Moon Orbit')
plt.plot(x0, y0, label='Earth Orbit')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Moon\'s Orbit around the Earth')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```
